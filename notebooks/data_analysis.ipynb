```python
import json
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
import os
```

```python
fnames = [fname for fname in os.listdir('../data/') if fname.endswith('.json')]

datas = [json.load(open("../data/" + fname, "r")) for fname in fnames]
```

# Tokenized Lengths

```python
lengths = []

for data in datas:
    for sample in data:
        length = len(sample['tags'])
        lengths.append(length)
```

```python
sns.boxplot(lengths);
```

```python
sns.distplot(lengths);
```

```python
min(lengths), max(lengths), np.mean(lengths), np.std(lengths)
```

```python
sum(1 for length in lengths if length > 75) / len(lengths) * 100
```
